```python
import sounddevice as sd
import matplotlib.pyplot as plt
from scipy.io import wavfile as wav
from scipy.fft import fft
import numpy as np
```

```python
rate, x = wav.read('./sounds/recording0.wav')
plt.plot(x)
plt.text(0-44100, np.max(x), 's', fontsize=16,
         bbox=dict(facecolor='red', alpha=1))
plt.show()

```

```python
x_n = x / np.max(x) #normalizaci√≥n
fft_x = np.fft.fft(x)
fft_x = abs(fft_x)
freq=np.arange(0,len(fft_x))

plt.figure()
plt.plot(freq, fft_x, color='blue')
plt.xlabel('Freq (in kHz)')
plt.ylabel('Amplitude')
plt.show()
```
